```python
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
import joblib
from ucimlrepo import fetch_ucirepo  

# ---- 1️⃣ Load the Dataset ----
print("📥 Fetching dataset from UCI ML Repository...")
heart_disease = fetch_ucirepo(id=45)  

X = heart_disease.data.features
y = heart_disease.data.targets

print("\n📊 Class Distribution Before Balancing:")
print(y.value_counts())

# ---- 2️⃣ Handle Missing Values ----
print("\n🛠 Handling missing values...")
X.fillna(X.median(), inplace=True)  # Fill missing numerical values with median
y.fillna(y.mode()[0], inplace=True)  # Fill missing target values with mode

# ---- 3️⃣ Apply Controlled SMOTE (Avoid Overfitting) ----
print("\n🔄 Applying SMOTE...")
smote = SMOTE(sampling_strategy={1: 100, 2: 100, 3: 100, 4: 100}, random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

print("\n✅ Class Distribution After SMOTE Resampling:")
print(pd.Series(y_resampled).value_counts())

# ---- 4️⃣ Normalize Data (Essential for Accuracy) ----
scaler = StandardScaler()
X_resampled_scaled = scaler.fit_transform(X_resampled)

# ---- 5️⃣ Train-Test Split ----
X_train, X_test, y_train, y_test = train_test_split(X_resampled_scaled, y_resampled, test_size=0.2, random_state=42)

# ---- 6️⃣ Train an Improved RandomForest Model ----
print("\n🤖 Training the Improved RandomForest model...")
model = RandomForestClassifier(n_estimators=200, max_depth=20, random_state=42)
model.fit(X_train, y_train)

# ---- 7️⃣ Save the Model and Feature Names ----
model_path = "/Users/fatimaiqbal/heart_disease/heart_disease_model_balanced.pkl"
os.makedirs(os.path.dirname(model_path), exist_ok=True)

joblib.dump((model, scaler, X.columns.tolist()), model_path)
print(f"\n✅ Model Retrained and Saved at: {model_path}")
```
